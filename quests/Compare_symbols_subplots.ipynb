코스피와 관심종목 비교 Charts 작성
- figuresize = (4,1)
- 비교 컬럼은 자유

```python
# 코스피 정보 출력

import matplotlib.pyplot as plt

import matplotlib as mpl 
# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'

import FinanceDataReader as fdr

kospi = fdr.DataReader('KS11', '2021-08-10')
krafton = fdr.DataReader('259960', '2021-08-10')
```

```python
plt.plot(kospi['Change'], label='코스피 변동률')
plt.plot(krafton['Change'], label='크래프톤 변동률')


plt.title('크래프톤 상장 후 코스피와의 비교')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.show()
```

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 10))  # figsize로 전체 그래프 크기 조절



# 첫 번째 subplot - 코스피 변동률
axes[0].plot(kospi['Change'], label='코스피 변동률')
axes[0].legend()

# 두 번째 subplot - 크래프톤 변동률
axes[1].plot(krafton['Change'], label='크래프톤 변동률')
axes[1].legend()

# 세 번째 subplot - 두 변동률 차이
axes[2].plot(kospi['Change'] - krafton['Change'], label='변동률 차이')

# 네 번째 subplot - 두 변동률 비교
axes[3].plot(kospi['Change'], label='코스피 변동률')
axes[3].plot(krafton['Change'], label='크래프톤 변동률')
axes[3].legend()



plt.tight_layout()  # subplot 간 간격 자동 조절
```
