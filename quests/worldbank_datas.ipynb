```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl 
# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
import seaborn as sns
```

```python


CO2EmissionRate = wbdata.get_data(indicator='CC.CO2.EMSE.EN', country=['USA', 'CHN', 'IND'], date=('2010', '2020'))
dfCO2EmissionRate = pd.DataFrame(CO2EmissionRate)
dfCO2EmissionRate[['country', 'date', 'value']]
plt.figure(figsize=(12, 6))
# plt.fill_between(data=dfCO2EmissionRate, x='date', y='value', hue='countryiso3code')
plt.title('Unemployment Rate Comparison by Country (2010-2020)')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
unemployment = wbdata.get_data(indicator='SL.UEM.TOTL.ZS', country=['DEU', 'GBR', 'FRA'], date=('2010', '2020'))

# dfUnemployment = pd.DataFrame(unemployment)
# # dfUnemployment[['countryiso3code', 'date', 'value']]
# # # dfUnemployment.describe()
# # dfUnemployment['countryiso3code	', 'date', 'value']
# # # plt.plot(finalUnem)
# 방법 1: 원하는 칼럼만 대괄호로 선택
dfUnemployment = pd.DataFrame(unemployment)[['countryiso3code', 'date', 'value']]
plt.figure(figsize=(12, 6))
sns.barplot(data=dfUnemployment, x='date', y='value', hue='countryiso3code')
plt.title('Unemployment Rate Comparison by Country (2010-2020)')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# # 방법 2: 원하는 칼럼들을 리스트로 지정
# columns_to_select = ['countryiso3code', 'date', 'value']
# dfUnemployment = pd.DataFrame(unemployment)[columns_to_select]


```

```python
internetUsers = wbdata.get_data(indicator='IT.NET.USER.ZS', country=['USA', 'CHN', 'BRA'], date=('2010', '2020'))

dfinternetUsers = pd.DataFrame(internetUsers)[['countryiso3code', 'date', 'value']]
plt.figure(figsize=(12, 6))
sns.barplot(data=dfinternetUsers, x='date', y='value', hue='countryiso3code')
plt.title('Internet Use Rate Comparison by Country (2010-2020)')
plt.xlabel('Year')
plt.ylabel('Internet Use Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()




```

```python
eduRate = wbdata.get_data(indicator='SE.XPD.TOTL.GD.ZS', country=['USA', 'GBR', 'DEU'], date=('2010', '2019'))


dfeduRate = pd.DataFrame(eduRate)[['countryiso3code', 'date', 'value']]
plt.figure(figsize=(12, 6))
sns.barplot(data=dfeduRate, x='date', y='value', hue='countryiso3code')
plt.title('Education Expenditure Level Comparison by Country (2010-2019)')
plt.xlabel('Year')
plt.ylabel('Internet Use Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
