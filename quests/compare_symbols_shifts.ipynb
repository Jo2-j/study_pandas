- 주식 지수와 관심종목들 비교 
- 1일, 5일, 10일 전 등락률 비교
- Charts 작성
- 등락률 = (오늘종가-전 종가 / 전 종가) * 100

```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import matplotlib as mpl 
# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
fdrSNP = fdr.DataReader('S&P500', '2024-11-01')

fdrSENS = fdr.DataReader('SENS', '2024-11-01')

fdrSNP
# fdrSENS.info()
```

```python
fdrSNP['전일종가'] = fdrSNP['Adj Close'].shift(periods=1)
fdrSNP['5일전 종가'] = fdrSNP['Adj Close'].shift(periods=5)
fdrSNP['10일전 종가'] = fdrSNP['Adj Close'].shift(periods=10)
fdrSNP['등락율'] = ((fdrSNP['Adj Close']- fdrSNP['전일종가']) / fdrSNP['전일종가']) * 100

fdrSNP[['Volume','Adj Close', '전일종가', '5일전 종가', '10일전 종가', '등락율']]

```

```python
fdrSENS['전일종가'] = fdrSENS['Adj Close'].shift(periods=1)
fdrSENS['5일전 종가'] = fdrSENS['Adj Close'].shift(periods=5)
fdrSENS['10일전 종가'] = fdrSENS['Adj Close'].shift(periods=10)
fdrSENS['등락율'] = ((fdrSENS['Adj Close']- fdrSENS['전일종가']) / fdrSENS['전일종가']) * 100

fdrSENS[['Volume','Adj Close', '전일종가', '5일전 종가', '10일전 종가', '등락율']]


```

```python
plt.plot(fdrSENS['등락율'], label='SENS')
plt.plot(fdrSNP['등락율'], label='S&P')


plt.title('2024년 11월이후 지수비교')
plt.xlabel('Date')
plt.ylabel('Change Rate, %')
plt.legend()
plt.tight_layout()
plt.show()
```

```python

plt.plot(fdrSNP['전일종가'], label='1 Day')
plt.plot(fdrSNP['5일전 종가'], label='5 Days')
plt.plot(fdrSNP['10일전 종가'], label='10 Days')


plt.title('S&P 2024년 11월이후 종가비교')
plt.xlabel('Date')
plt.ylabel('Close, $')
plt.legend()
plt.tight_layout()
plt.show()
```

```python

plt.plot(fdrSENS['전일종가'], label='1 Day')
plt.plot(fdrSENS['5일전 종가'], label='5 Days')
plt.plot(fdrSENS['10일전 종가'], label='10 Days')


plt.title('SENS 2024년 11월이후 종가비교')
plt.xlabel('Date')
plt.ylabel('Close, $')
plt.legend()
plt.tight_layout()
plt.show()
```
