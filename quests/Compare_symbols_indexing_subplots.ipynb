```python
# 코스피 정보 출력

import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl 
# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'

import FinanceDataReader as fdr

```

```python
# 한국 지수
kosKOR = fdr.DataReader(symbol='KS11', start='1990', end='2022')  # KOSPI: 940 종목


kosdaqKOR = fdr.DataReader(symbol='KQ11', start='1990', end='2022')  # KOSDAQ: 1,597 종목


# 미국 지수
dowUS = fdr.DataReader(symbol='DJI', start='1990', end='2022')  # 다우존스 지수


nasdUS = fdr.DataReader(symbol='IXIC', start='1990', end='2022')  # 나스닥 종합지수

snpUS = fdr.DataReader(symbol='S&P500', start='1990', end='2022')  # S&P500 지수

rutUS = fdr.DataReader(symbol='RUT', start='1990', end='2022')  # 러셀2000 지수

vixUS = fdr.DataReader(symbol='VIX', start='1990', end='2022')  # VIX지수

# 아시아 지수
ssecCH = fdr.DataReader(symbol='SSEC', start='1990', end='2022')  # 상해 종합지수

hsiHK = fdr.DataReader(symbol='HSI', start='1990', end='2022')  # 항셍지수

nikkeiJP = fdr.DataReader(symbol='N225', start='1990', end='2022')  # 일본 닛케이지수

# 유럽 지수
ftseUK = fdr.DataReader(symbol='FTSE', start='1990', end='2022')  # 영국 FTSE100

cacFR = fdr.DataReader(symbol='FCHI', start='1990', end='2022')  # 프랑스 CAC40

daxGER = fdr.DataReader(symbol='GDAXI', start='1990', end='2022')  # 독일 DAX30



# # Global Market listings 글로벌 시장 거래소별 전종목 리스팅
# sseCH = fdr.StockListing('SSE') # 상하이 증권거래소 (Shanghai Stock Exchange: SSE): 1천+ 종목
# szCH = fdr.StockListing('SZSE') # 선전 증권거래소(Shenzhen Stock Exchange: SZSE): 1천+ 종목
# hkexHK = fdr.StockListing('HKEX') # 홍콩 증권거래소(Hong Kong Exchange: HKEX): 2천5백+ 종목
# tseJP = fdr.StockListing('TSE') # 도쿄 증권거래소(Tokyo Stock Exchange: TSE): 3천9백+ 종목
# hoseVI = fdr.StockListing('HOSE') # 호찌민 증권거래소(Ho Chi Minh City Stock Exchange: HOSE): 4백+ 종목
```

```python
seriesKosKOR = (kosKOR['Close'] / kosKOR['Close'].iloc[0]) * 100
seriesKosdaqKOR = (kosdaqKOR['Close'] / kosdaqKOR['Close'].iloc[0]) * 100
seriesDowUS = (dowUS['Close'] / dowUS['Close'].iloc[0]) * 100
seriesNasdUS = (nasdUS['Close'] / nasdUS['Close'].iloc[0]) * 100
seriesSnpUS = (snpUS['Close'] / snpUS['Close'].iloc[0]) * 100
seriesRutUS = (rutUS['Close'] / rutUS['Close'].iloc[0]) * 100
seriesVixUS = (vixUS['Close'] / vixUS['Close'].iloc[0]) * 100
seriesSsecCH = (ssecCH['Close'] / ssecCH['Close'].iloc[0]) * 100
seriesHsiHK = (hsiHK['Close'] / hsiHK['Close'].iloc[0]) * 100
seriesNikkeiJP = (nikkeiJP['Close'] / nikkeiJP['Close'].iloc[0]) * 100
seriesFtseUK = (ftseUK['Close'] / ftseUK['Close'].iloc[0]) * 100
seriesCacFR = (cacFR['Close'] / cacFR['Close'].iloc[0]) * 100
seriesDaxGER = (daxGER['Close'] / daxGER['Close'].iloc[0]) * 100
```

```python
# seriesDowUS.describe()
# seriesDowUS.info()
```

```python
plt.plot(kosKOR.index , seriesKosKOR, label='KOSPI')
plt.plot(dowUS.index , seriesDowUS, label='DOW')
plt.plot(nasdUS.index , seriesNasdUS, label='NASDAQ')
plt.plot(snpUS.index , seriesSnpUS, label='S&P500')
plt.plot(rutUS.index , seriesRutUS, label='RUSSELL2000')
plt.plot(vixUS.index , seriesVixUS, label='VIX')
plt.plot(ssecCH.index , seriesSsecCH, label='SSEC')
plt.plot(hsiHK.index , seriesHsiHK, label='HSI')
plt.plot(nikkeiJP.index , seriesNikkeiJP, label='NIKKEI225')
plt.plot(ftseUK.index , seriesFtseUK, label='FTSE100')
plt.plot(cacFR.index , seriesCacFR, label='CAC40')
plt.plot(daxGER.index , seriesDaxGER, label='DAX30')
plt.title('지수화 비교')
plt.legend()
plt.show()
```
