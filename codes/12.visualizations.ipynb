### 한글처리

```python
import matplotlib.pyplot as plt

import matplotlib as mpl 
# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
plt.plot(x_data, y_data, label='x_data_legend')
plt.plot(y_data, x_data, label='y_data_legend')

plt.title('Visualization With Title')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend()
plt.show()
```

```python
plt.plot(x_data, y_data, label='x_data_범례')
plt.plot(y_data, x_data, label='y_data_범례')

plt.title('Visualization With 타이틀')
plt.xlabel('X 데이터')
plt.ylabel('Y 데이터')
plt.legend()
plt.savefig('visualizations_plot.png') # show 전에 해야함
plt.show()
# plt.savefig('visualizations_plot.png')
```

```python
df_loan_remain_sorted
```

```python
plt.plot(df_loan_remain_sorted.index, df_loan_remain_sorted['전체'], label='전체', marker='o')
plt.plot(df_loan_remain_sorted.index, df_loan_remain_sorted['1금융권'], label='1금융권', marker='x')
plt.plot(df_loan_remain_sorted.index, df_loan_remain_sorted['2금융권'], label='2금융권', marker='s')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
