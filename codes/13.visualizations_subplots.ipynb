### Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 데이터가 맞아야 하는 이유 x,y 에 교차하는 점을 찍는거니까

x_data_first = [1,2,3,4]
y_data_first = [1,2,3,4]

x_data_second = [2,3,1,1]
y_data_second = [2,1,4,1]

x_data_third = [1,2,3,4]
y_data_third = [2,1,4,1]

x_data_fourth = [4,3,2,1]
y_data_fourth = [4,3,2,1]
```

## 정렬

###### add_subplot, subplots()

```python
# fig, axes =plt.subplots(2,2)
# fig, axes =plt.subplots(2,2, figsize=(14,6))
fig, axes =plt.subplots(1,2)
```

```python
type(axes), axes.shape, type(axes[0]) 
```

```python
fig, axes =plt.subplots(1,2)
axes[0].plot(x_data_first, y_data_first)
axes[1].plot(x_data_third, y_data_third)
plt.show()
```

### 구성요소 추가

```python
#### 순간 quests : 1,4 그리드 그래프 그리기
fig, axes =plt.subplots(1,4)

fig.suptitle('Super Title')

axes[0].plot(x_data_first, y_data_first)

axes[1].plot(x_data_third, y_data_third)
axes[1].set_title('First')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')



axes[2].plot(x_data_second, y_data_second)
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')


axes[3].plot(x_data_fourth, y_data_fourth)
axes[3].set_xlabel('x')
axes[3].legend()

plt.tight_layout()
plt.show()
```
